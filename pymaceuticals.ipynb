```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
data_frame = pd.merge(study_results, mouse_metadata, how="outer", on="Mouse ID")

# Display the data table for preview
data_frame.head()
```

```python
# Checking the number of mice.
data_frame["Mouse ID"].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_df = data_frame[data_frame.duplicated(subset=["Mouse ID","Timepoint"])]
duplicate_IDs = duplicate_df["Mouse ID"].unique()
duplicate_IDs
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
delete_df = data_frame[data_frame["Mouse ID"]=="g989"]
delete_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
data_frame = data_frame.drop(delete_df.index)
data_frame.head()
```

```python
# Checking the number of mice in the clean DataFrame.
data_frame["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume.
data_mean = data_frame.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
summary_statistics = pd.DataFrame(data_mean).rename(columns={"Tumor Volume (mm3)": "Mean Tumor Volume"})

data_median = data_frame.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
summary_statistics["Median Tumor Volume"] = data_median

data_variance = data_frame.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
summary_statistics["Tumor Volume Variance"] = data_variance

data_stddev = data_frame.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
summary_statistics["Tumor Volume Std. Dev."] = data_stddev

data_stderr = data_frame.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()
summary_statistics["Tumor Volume Std. Err."] = data_stderr

# Assemble the resulting series into a single summary DataFrame.
summary_statistics
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# # Using the aggregation method, produce the same summary statistics in a single line
summary_statistics2 = data_frame.groupby(["Drug Regimen"])[["Tumor Volume (mm3)"]].agg(['mean', 'median', 'var', 'std', 'sem'])
summary_statistics2
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = data_frame.groupby('Mouse ID')['Timepoint'].max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_volume = pd.merge(last_timepoint, data_frame, on=['Mouse ID', 'Timepoint'], how='left')
final_tumor_volume.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_data = final_tumor_volume.loc[final_tumor_volume['Drug Regimen'] == drug, 'Tumor Volume (mm3)']
    
    # add subset
    tumor_vol_data.append(tumor_data)
    
    # Determine outliers using upper and lower bounds
    quartiles = tumor_data.quantile([.25, .5, .75])
    lowerq = quartiles[.25]
    upperq = quartiles[.75]
    IQR = upperq - lowerq
    lower_bound = lowerq - (1.5 * IQR)
    upper_bound = upperq + (1.5 * IQR)
    outliers = tumor_data.loc[(tumor_data < lower_bound) | (tumor_data > upper_bound)]

    print(f"{drug}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig1, ax1 = plt.subplots()
ax1.set_title('Distrubution of the tumor volume for each treatment group')
ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.boxplot(tumor_vol_data, labels=treatments, flierprops=dict(marker='o', color='black', markersize=12, markerfacecolor='red'))
plt.show()
```

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
Capomulin_data = data_frame.loc[data_frame['Drug Regimen'] == 'Capomulin']
Capomulin_MouseIDs = Capomulin_data['Mouse ID'].value_counts()
mouse_id = Capomulin_MouseIDs.index[0]  # Selecting the mouse ID with highest number of time points
single_mouse_data = Capomulin_data[Capomulin_data['Mouse ID'] == mouse_id]
plt.plot(single_mouse_data['Timepoint'], single_mouse_data['Tumor Volume (mm3)'])
plt.title(f"Capomulin treatment of mouse {mouse_id}")
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
avg_tumor_volume = Capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
weights = Capomulin_data.groupby('Mouse ID')['Weight (g)'].first()
plt.scatter(weights, avg_tumor_volume)
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(weights, avg_tumor_volume)
print(f'The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}')

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(weights, avg_tumor_volume)

plt.scatter(weights, avg_tumor_volume)
plt.plot(weights, intercept + slope * weights, 'r')
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

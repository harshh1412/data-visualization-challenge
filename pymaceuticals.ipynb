```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
data_frame = pd.merge(study_results, mouse_metadata, how="outer", on="Mouse ID")

# Display the data table for preview
data_frame.head()
```

```python
# Checking the number of mice.
data_frame["Mouse ID"].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_df = data_frame[data_frame.duplicated(subset=["Mouse ID","Timepoint"])]
duplicate_IDs = duplicate_df["Mouse ID"].unique()
duplicate_IDs
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
delete_df = data_frame[data_frame["Mouse ID"]=="g989"]
delete_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
data_frame = data_frame.drop(data_frame[data_frame["Mouse ID"]=="g989"].index)
data_frame.head()
```

```python
# Checking the number of mice in the clean DataFrame.
data_frame["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume.
data_mean = data_frame.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
summary_statistics = pd.DataFrame(data_mean).rename(columns={"Tumor Volume (mm3)": "Mean Tumor Volume"})

data_median = data_frame.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
summary_statistics["Median Tumor Volume"] = data_median

data_variance = data_frame.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
summary_statistics["Tumor Volume Variance"] = data_variance

data_stddev = data_frame.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
summary_statistics["Tumor Volume Std. Dev."] = data_stddev

data_stderr = data_frame.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()
summary_statistics["Tumor Volume Std. Err."] = data_stderr

# Assemble the resulting series into a single summary DataFrame.
summary_statistics
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas.
data_bar = data_frame["Drug Regimen"].value_counts()
data_frame_bar = pd.DataFrame(data_bar)

data_frame_bar.plot(kind="bar", title="Observations per Drug Regimen" , xlabel= "Drug Regimen", ylabel="# of Observed Mouse Timepoints")
plt.show()
```

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using pyplot.
data_bar_reset = data_bar.reset_index(["Drug Regimen"])
plt.bar(data_bar_reset["Drug Regimen"], data_bar_reset["count"])
plt.xticks(rotation= "vertical")
plt.title("Observations per Drug Regimen")
plt.xlabel("Drug Regimen")
plt.ylabel("# of Observed Mouse Timepoints")
plt.show()
```
